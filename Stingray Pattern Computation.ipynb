```python
# SageMath code: list Lambda-coordinates for all partitions λ
# with length(λ) ≤ r, e-weight = w.

from sage.combinat.partition import Partitions, Partition

def lambda_plus_rho_coords(lam, r):
    """
    Input:
        lam : an iterable of parts of λ (e.g. (5,3,1))
        r   : positive integer (rank = r)
    Output:
        tuple (Λ1,...,Λ_{r-1}) for (λ+ρ) in type A_{r-1}.

    Convention:
    Λ_i = fundamental weights of sl_r.
    Coordinates are λ_i - λ_{i+1} + 1, with λ padded by 0s to length r.
    """
    lam_padded = list(lam) + [0]*(r - len(lam))
    return tuple(lam_padded[i] - lam_padded[i+1] + 1 for i in range(r-1))


def e_core_and_weight(lam, e):
    """
    Input:
        lam : tuple/list of parts of λ
        e   : modulus
    Output:
        (core_e(λ), e_weight(λ)) where
        e_weight(λ) = (|λ| - |core_e(λ)|)/e  (integer)
    Uses Sage's built-ins.
    """
    P = Partition(lam)
    core = P.core(e)
    lam_size = sum(lam)
    core_size = sum(core)
    w_val = (lam_size - core_size) // e
    return core, int(w_val)


def lambdas_fixed_e_weight(r, e, w):
    """
    Input:
        r : max number of parts (so length(λ) ≤ r)
        e : positive integer (modulus for e-core)
        w : desired e-weight

    Output:
        list of all partitions λ (as Partition objects)
        such that length(λ) ≤ r and e-weight(λ) = w.
    """
    # Upper bound on size of λ:
    # If core has size ≥0, then |λ| = |core| + e*w.
    # The minimal possible |core| is 0, so |λ| ≤ e*w + something.
    # We don't know a priori which cores appear, but |λ| must be ≥ e*w
    # and not explode too far above e*w. We'll take a safe search range.
    #
    # A simple safe bound is: search all partitions of size up to e*w + e*r.
    # This is crude but finite. You can tune if needed.
    max_size = e*w + e*r

    res = []
    for n in range(e*w, max_size+1):
        for lam in Partitions(n):
            if len(lam) <= r:
                _, w_val = e_core_and_weight(lam, e)
                if w_val == w:
                    res.append(Partition(lam))
    return res


def Lambda_coordinates_for_all(r, e, w):
    """
    Master function.
    Returns a list of dicts, each dict containing:
        'lambda'  : the partition λ as a tuple
        'Lambda'  : (Λ1,...,Λ_{r-1}) coords of λ+ρ
        'ecore'   : the e-core of λ
        'eweight' : the e-weight of λ (should all be w)
    """
    ans = []
    for lam in lambdas_fixed_e_weight(r, e, w):
        coords = lambda_plus_rho_coords(lam, r)
        core, w_val = e_core_and_weight(lam, e)
        ans.append({
            'lambda'  : tuple(lam),
            'Lambda'  : coords,
            'ecore'   : tuple(core),
            'eweight' : w_val,
        })
    return ans
def Lambda_data(r, e, w):
    """
    Return rich data for each λ.
    """
    out = []
    for lam in lambdas_fixed_e_weight(r, e, w):
        coords = lambda_plus_rho_coords(lam, r)
        core, w_val = e_core_and_weight(lam, e)
        out.append({
            'lambda'  : tuple(lam),
            'Lambda'  : coords,
            'ecore'   : tuple(core),
            'eweight' : w_val,
        })
    return out
def Lambda_coord_list(r, e, w):
    """
    Return a sorted list of distinct Λ-coordinate tuples only.
    """
    data = Lambda_data(r, e, w)
    coord_set = { item['Lambda'] for item in data }  # unique tuples
    # sort for reproducibility (lexicographic)
    return sorted(coord_set)

#########################################
# Example usage:
#########################################

r  = 4   # at most 4 parts, so sl_4
e  = 10   # modulus
w  = 0  # e-weight

data = Lambda_coordinates_for_all(r, e, w)

# Pretty print
# for item in data:
#     print("λ =", item['lambda'],
#           "  Λ-coords(λ+ρ) =", item['Lambda'],
#           "  e-core =", item['ecore'],
#           "  e-weight =", item['eweight'])
print(Lambda_coord_list(r, e, w))
```
