```python
from sage.combinat.partition import Partition, Partitions

def omega_coords_lambda_plus_rho(lam, r):
    """
    Return (ω1, …, ω_{r-1}) for λ+ρ in type A_{r-1},
    i.e. ω_i = λ_i - λ_{i+1} + 1 with λ padded by zeros to length r.
    """
    lam = list(lam) + [0]*(r - len(lam))
    return tuple(lam[i] - lam[i+1] + 1 for i in range(r-1))

# e_core_and_weight is no longer needed, we will inline it.

def dominant_weights_sl_r(r, e, w, include_empty=True):
    """
    Compute the set of dominant weights ω(λ+ρ) arising from partitions λ
    with length(λ) ≤ r and e-weight(λ) = w.
    
    This version uses a rigorous upper bound on the size n = |λ|
    to guarantee all partitions are found.
    """
    S = set()
    start_n = 0 if (w == 0 and include_empty) else 1

    # 1. Calculate the rigorous upper bound for the core size
    # C_max = (e-1) * r*(r+1) / 2
    max_core_size = (e - 1) * r * (r + 1) // 2
    
    # 2. Calculate the maximum partition size n = |λ| to check
    # N_max = C_max + e * w
    N_lambda_max = max_core_size + e * w

    # Optional: print the bound to see what's happening
    # print(f"r={r}, e={e}, w={w} | Max core size: {max_core_size}, Max |λ|: {N_lambda_max}")

    # 3. Iterate from n=start_n up to the guaranteed bound
    for n in range(start_n, N_lambda_max + 1):
        for lam in Partitions(n, max_length=r):
            
            # --- Inlined e_core_and_weight logic ---
            P = Partition(lam)
            core = P.core(e)
            core_size = sum(core)
            
            # Check 1: Divisibility. |λ| and |core| must be congruent mod e.
            if (n - core_size) % e != 0:
                continue
                
            # Check 2: e-weight
            w_lam = (n - core_size) // e
            if w_lam != w:
                continue
                
            # Check 3: Core length (redundant, as len(core) <= len(lam) <= r)
            # But we keep it to match your original function's explicit logic.
            # if len(core) > r:
            #     continue
            
            # If all checks pass, compute and add the weight
            S.add(omega_coords_lambda_plus_rho(lam, r))

    return tuple(sorted(S))

# -----------------------------
# r=4;
# for e in range(2,10):
#     for w in range(0,10):
#         print("e=",e,"w=",w,len(dominant_weights_sl_r(r, e, w)))

# Heavier case (works fine on a modern laptop; tune max_part upward if needed):
# W = dominant_weights_sl_r(r=3, e=30, w=30, max_part=100)
# print(len(W))
```

```python
from sage.combinat.partition import Partition, Partitions

# ---------------- Your function (lightly optimized) ----------------
def omega_coords_lambda_plus_rho(lam, r):
    """
    Return (ω1, …, ω_{r-1}) for λ+ρ in type A_{r-1},
    i.e. ω_i = λ_i - λ_{i+1} + 1 with λ padded by zeros to length r.
    """
    lam = list(lam) + [0]*(r - len(lam))
    return tuple(lam[i] - lam[i+1] + 1 for i in range(r-1))

def dominant_weights_sl_r(r, e, w, include_empty=True):
    """
    Compute the set of dominant weights ω(λ+ρ) arising from partitions λ
    with length(λ) ≤ r and e-weight(λ) = w.
    """
    S = set()
    start_n = 0 if (w == 0 and include_empty) else 1

    # 1) core size bound & |λ| bound (same as yours)
    max_core_size = (e - 1) * r * (r + 1) // 2
    N_lambda_max = max_core_size + e * w

    # --- micro speed-ups via local bindings ---
    Partitions_n = Partitions
    PartitionObj = Partition
    omega_fn = omega_coords_lambda_plus_rho

    # 2) key improvement: skip impossible n < e*w
    for n in range(max(start_n, e*w), N_lambda_max + 1):
        for lam in Partitions_n(n, max_length=r):
            P = PartitionObj(lam)
            core = P.core(e)
            core_size = sum(core)
            # divisibility + exact e-weight
            dn = n - core_size
            if dn % e != 0:
                continue
            if dn // e != w:
                continue
            S.add(omega_fn(lam, r))

    return tuple(sorted(S))

# ---------------- Parallel driver (over (e, w)) ----------------
NCORES = 6  # <- set to your physical cores (or a bit less)

@parallel(NCORES)
def count_for_e_w(e, w, r, include_empty):
    try:
        return len(dominant_weights_sl_r(r, e, w, include_empty=include_empty))
    except Exception as ex:
        # return a structured error instead of a bare string
        return {"error": f"{type(ex).__name__}: {ex}"}

def _normalize_args(a):
    # Sage @parallel sometimes gives args as ((... ),) instead of (...)
    if isinstance(a, tuple) and len(a) == 1 and isinstance(a[0], tuple):
        return a[0]
    return a

# ---------------- Run exactly your grid and print ----------------
r = 5
include_empty = True
e_vals = range(2, 11)
w_vals = range(0, 11)

jobs = [(e, w, r, include_empty) for e in e_vals for w in w_vals]

ok, err = [], []
for (args, out) in count_for_e_w(jobs):
    pargs = _normalize_args(args)   # (e, w, r, include_empty)
    if isinstance(out, dict) and "error" in out:
        err.append((pargs, out["error"]))
    else:
        ok.append((pargs, int(out)))

# show any worker errors clearly
for (e, w, r_, inc_), msg in err:
    print(f"[ERROR] e={e} w={w}: {msg}")

# stable ordered output, exactly like you wanted
for (e, w, r_, inc_), c in sorted(ok, key=lambda t: (t[0][0], t[0][1])):
    print("e=", e, "w=", w, c)
```
